```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(filename, skiprows=1, delimiter=","):
    """
    Function loads data stored in the file filename and returns it as a numpy ndarray.
    
    Inputs:
        filename: given as a string.
        
    Outputs:
        Data contained in the file, returned as a numpy ndarray
    """
    return np.loadtxt(filename, skiprows=skiprows, delimiter=delimiter)
```

```python
def load_movies_data():
    """
    Load the movies.txt file.
    """
    genre_index_start = -19
    movies_data = {}
    movies_file = open("data/movies.txt", "r")
    lines = movies_file.readlines()
    for i in range(len(lines)):
        line = lines[i].split()
        length = len(line)
        movies_data[int(line[0])] = \
            [" ".join(line[1:length + genre_index_start - 1])] \
            + list(map(int, line[genre_index_start:]))
    return movies_data
```

```python
data = load_data("data/data.txt", skiprows=0, delimiter="\t")
train_data = load_data("data/train.txt", skiprows=0, delimiter="\t")
test_data = load_data("data/test.txt", skiprows=0, delimiter="\t")
movies_data = load_movies_data()
```

## Removing Movies With Fewer Than 2 Ratings From Data

```python
last_movie_id = max(movies_data)
movie_num_rated_array = np.zeros(last_movie_id+1)

for i in range(len(data)):
    movie_num_rated_array[int(data[i,1])] += 1
```

```python
movies_to_keep = np.where(movie_num_rated_array > 1)[0]
```

```python
len(movies_to_keep)
```

```python
movie_id_map = {}
displacement = 0
exclude_id = -1
for movie_id in movies_data:
    if movie_id in movies_to_keep:
        movie_id_map[movie_id] = movie_id - displacement
    else:
        movie_id_map[movie_id] = exclude_id
        displacement += 1
```

```python
for i in movie_id_map:
    if movie_id_map[i] != i:
        if movie_id_map[i] != exclude_id:
            movies_data[movie_id_map[i]] = movies_data[i]
        del movies_data[i]
```

```python
new_data = []
for i in range(len(data)):
    movie_id = data[i,1]
    new_movie_id = movie_id_map[movie_id]
    if new_movie_id != exclude_id:
        new_data.append([data[i,0], new_movie_id, data[i,2]])
new_data = np.array(new_data)

new_train_data = []
for i in range(len(train_data)):
    movie_id = train_data[i,1]
    new_movie_id = movie_id_map[movie_id]
    if new_movie_id != exclude_id:
        new_train_data.append([train_data[i,0], new_movie_id,
                               train_data[i,2]])
new_train_data = np.array(new_train_data)

new_test_data = []
for i in range(len(test_data)):
    movie_id = test_data[i,1]
    new_movie_id = movie_id_map[movie_id]
    if new_movie_id != exclude_id:
        new_test_data.append([test_data[i,0], new_movie_id,
                              test_data[i,2]])
new_test_data = np.array(new_test_data)
```

```python
data = new_data
train_data = new_train_data
test_data = new_test_data
```

# First two collumns projection of U and V

```python
def proj_matrices(UT, V):
    # print(np.shape(UT))
    # print(np.shape(V))
    for i in range(len(V)):
        mean = np.mean(V[i])
        std = np.std(V[i])
        V[i] = (V[i] - mean)/std
        UT[:,i] = (UT[:,i] - mean)/std
    
    A, Sig, BT = np.linalg.svd(V)
    best2A = np.transpose(A[:,:2])
    # print(np.shape(best2A))
    UT_tild = np.transpose(np.dot((best2A), np.transpose(UT)))
    V_tild = np.dot((best2A), V)
    return UT_tild, V_tild
```

```python
def findUsers_original(movieIndices, data):
    users_found = []
    user_data = []
    user_average_rating_dict = {}
    for i, triple in enumerate(data):
        user = data[i][0]
        movie_id = data[i][1]
        rating = data[i][2]
        if(movie_id in movieIndices):
            if user not in users_found:
                users_found.append(user)
                user_average_rating_dict[user] = [rating, rating]
                user_data.append([user, movie_id, rating])
            else:
                user_average_rating_dict[user].append(rating)
                user_average_rating_dict[user][0] = \
                np.mean(np.array[user_average_rating_dict[user][1:]])
    return user_data
```

```python
def findUsers(movieIndices, data):
    users_found = []
    user_data = []
    user_average_rating_dict = {}
    for i, triple in enumerate(data):
        user = data[i][0]
        movie_id = data[i][1]
        rating = data[i][2]
        if(movie_id in movieIndices):
            if user not in users_found:
                users_found.append(user)
                user_average_rating_dict[user] = [rating, rating]
                user_data.append([user, movie_id, rating])
            else:
                user_average_rating_dict[user].append(rating)
                user_average_rating_dict[user][0] = \
                    np.mean(np.array(user_average_rating_dict[user][1:]))
    user_dict = {}
    for user in users_found:
        user_dict[user] = user_average_rating_dict[user][0]
    return user_dict
```

```python
def series_plot_U(UT_tild, user_dict, series_name, color="blue", continued=False,
                  fig=None, ax=None, plot_items=[], using_plot_items=False,
                  rating_range=None, title=""):
    
    if fig == None or ax == None:
        fig, ax = plt.subplots(figsize=(10,8))
    #print(str(len(UT_tild[0,series_movies])) + " " + series_name)
    Xs = []
    Ys = []
    titles = []
    colors = []
    
    if (not continued and len(plot_items) == 4) or \
        using_plot_items:
        Xs = plot_items[0]
        Ys = plot_items[1]
        titles = plot_items[2]
        colors = plot_items[3]
    
    for i in user_dict:
        idx = i
        idx = int(idx)
        x = UT_tild[idx-1][0]
        y = UT_tild[idx-1][1]
        rating = user_dict[i]
        
        if rating_range != None:
            if rating >= rating_range[0] and\
               rating <= rating_range[1]:
                Xs.append(x)
                Ys.append(y)
                # titles.append(str(i) + ": " + str(rating))  
                titles.append("")
                colors.append(color)
        else:
            Xs.append(x)
            Ys.append(y)
            # titles.append(str(i) + ": " + str(rating))
            titles.append("")
            colors.append(color)
    
    if not continued:
        ax.scatter(Xs, Ys, c=colors)
        # ax.set_yscale("symlog")
        ax.set_xlabel("dimension 1")
        ax.set_ylabel("dimension 2")

        for i in range(len(Xs)):
            ax.annotate(titles[i], (Xs[i], Ys[i]), ha="center")
    
    if continued:
        return fig, ax, [Xs, Ys, titles, colors]
    else:
        ax.set_title(title)
        fig.savefig(series_name + ".png")
```
