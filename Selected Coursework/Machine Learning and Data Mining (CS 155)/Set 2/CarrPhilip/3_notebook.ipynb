```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
n_alpha_lasso = 30
alpha_values_lasso = np.linspace(0.01, 3, n_alpha_lasso)
```

```python
# Lasso regression trained model weights vs alpha.
d = 9
W = np.zeros((n_alpha_lasso, d))
for i in range(len(alpha_values_lasso)):
    lasso = Lasso(alpha=alpha_values_lasso[i])
    lasso.fit(X_train, y_train)
    for j in range(d):
        W[i,j] = lasso.coef_[j]

plt.figure()
plt.title("Lasso regression trained model weights vs alpha")
plt.xlabel("alpha")
plt.ylabel("model weight")
for i in range(d):
    plt.plot(alpha_values_lasso, W[:, i], label="w%d" % (i+1))
plt.legend(ncol=2)
plt.savefig("3a_plot_lasso.png")
```

```python
alpha_values_ridge = np.arange(1, 1e4)
n_alpha_ridge = len(alpha_values_ridge)
```

```python
# Plot Ridge regression trained model weights vs alpha.
d = 9
W = np.zeros((n_alpha_ridge, d))
for i in range(len(alpha_values_ridge)):
    ridge = Ridge(alpha=alpha_values_ridge[i])
    ridge.fit(X_train, y_train)
    for j in range(d):
        W[i,j] = ridge.coef_[j]

plt.figure()
plt.title("Ridge regression trained model weights vs alpha")
plt.xlabel("alpha")
plt.ylabel("model weight")
for i in range(d):
    plt.plot(alpha_values_ridge, W[:, i], label="w%d" % (i+1))
plt.legend(ncol=2)
plt.savefig("3a_plot_ridge.png")
```
